## Redução de dimensionalidade de uma imagem:

Transformando uma imagem em tons de cinza e depois em preto e branco.

```python
# Referências:
# https://github.com/FlaysonSantos/Tratamento_de_img_com_python/blob/main/ImgPhyton.ipynb
```

```python
#Importando as bibliotecas:
from matplotlib import pyplot as plt
import matplotlib.image as mpimg



#lendo a imagem
img = mpimg.imread("Toulouse-Lautrec.jpg")


#Imprimindo a imagem
#Esta imagem foi criada por um algoritmo de inteligência artificial: Photoleap
plt.imshow(img, cmap='gray')

plt.show
```

```python
#Uma imagem é uma matriz tridimensional. A primeira camada se refere ao vermelho, a segunda ao verde e a terceira ao azul
#Lendo a primeira camada e salvando na variável Red

Red = img[:,:,0];


plt.imshow(Red, cmap='gray')

plt.show
```

```python
# Lendo a segunda camada e salvando na variável Green:

Green = img[:,:,1];

plt.imshow(Green, cmap='gray')

plt.show
```

```python
#Lendo a terceira camada e salvando na variável Blue:

Blue = img[:,:,2];

plt.imshow(Blue, cmap='gray')

plt.show
```

```python
# Se nós somarmos as 3 matrizes e dividirmos por 3, o resultado não vai ficar bom:

Gray = (Red + Green + Blue)/3

plt.imshow(Gray, cmap='gray')

plt.show
```

```python
#Para o resultado ficar bom, precisamos da um peso maior para o verde e menor para o vermelho e azul:

Gray2 = 0.2989 * Red + 0.5870 * Green + 0.1140 * Blue

plt.imshow(Gray2, cmap='gray')

plt.show
```

```python
#Este trecho de código mostra cada valor da matriz Gray2. Ele varia de 0 a 255

# tamanho = print(Gray2.shape)

# for i in range(len(Gray2)):
#     for j in range(len(Gray2[i])):

#       print(Gray2[i][j])

```

```python
# Este código percorre cada elemento da matriz da imagem. 
# Se o número for maior que 127, ele coloca 255.
# Se o número foi menor que 127, ele coloca 0.

for i in range(len(Gray2)):
    for j in range(len(Gray2[i])):

     # print(Gray[i][j])
        if(Gray2[i][j] < 127):
          Gray2[i][j] = 0;

        else:
          Gray2[i][j] = 255;



plt.imshow(Gray2, cmap='gray')

plt.show

```
